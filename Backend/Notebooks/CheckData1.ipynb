```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet("cleaned_data.parquet")
df.head()
```

```python
ffb = df['ffb_1%_oer']
ffb.head()
ffb.shape
```

```python
returns = ffb.pct_change().dropna() * 100

avg_daily_change = returns.mean()
volatility = returns.std()

print("Average daily change: ", avg_daily_change, "%")
print("Volatily (std dev): ", volatility, "%")
```

```python
import matplotlib.pyplot as plt

returns.hist(bins=50, edgecolor="black")
plt.title("Distribution of Daily % Changes")
plt.xlabel("Daily % Change")
plt.ylabel("Frequency")
plt.show()
```
