```python
import pandas as pd
import numpy as np
```

```python
excel_file="C:/Users/User/Desktop/SEM 7/FYP I/Dataset/Meh/full_weather.csv"
df = pd.read_csv(excel_file)
print(df.head())
```

```python
kuching_df=df[df['city'] == "Kuching"]
print(kuching_df)
```

```python
bintulu_df=df[df['city'] == "Bintulu"]
print(bintulu_df.head())
bintulu_df.shape
```

```python
# Convert datetime column to proper datetime type
bintulu_df["datetime"] = pd.to_datetime(bintulu_df["datetime"], errors="coerce")

# Filter by date range
mask = (bintulu_df["datetime"] >= "2019-01-01") & (bintulu_df["datetime"] <= "2024-08-31")
df_bintulu = bintulu_df.loc[mask]


# Extract date only (ignore time part)
bintulu_df["date"] = bintulu_df["datetime"].dt.date

# Group by date â†’ calculate daily average temperature & humidity
df_daily_avg = bintulu_df.groupby("date")[["temperature", "humidity"]].mean().reset_index()

# Save to Excel
output_file = "C:/Users/User/Desktop/SEM 7/FYP I/Dataset/bintulu_weather_daily_avg.xlsx"
df_daily_avg.to_excel(output_file, index=False)
```
